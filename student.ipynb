```python
def hist_plot(variable):
  variables=df[variable].value_counts()
  plt.figure(figsize=(9,3))
  plt.hist(variables.index,weights=variables.values)
  plt.xticks(variables.index,variables.index.values)
  plt.xlabel(variable)
  plt.ylabel('Frequency')
  plt.title(variable)
  plt.show()
  print("{}: \n {}".format(variables,variables))
```

```python
num_columns = df.select_dtypes(include=["int64", "float64"]).columns.tolist()
```

```python
num_columns
```

```python
for i in num_columns:
    hist_plot(i)
    plt.show()
```

```python
def bar_plot(variable):
    var=df[variable]
    varValue=var.value_counts()
    plt.figure(figsize=(9,3))
    plt.bar(varValue.index,varValue)
    plt.xticks(varValue.index,varValue.index.values)
    plt.ylabel("Frequency")
    plt.title(variable)
    plt.show()
    print("{}: \n {}".format(variable,varValue))
```

```python
for i in num_columns:
    bar_plot(i)
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming your DataFrame is already loaded as 'df'
# If 'Prog Code' is not a category, convert it
if 'Prog Code' in df.columns:
    if not pd.api.types.is_categorical_dtype(df['Prog Code']):
       df['Prog Code'] = df['Prog Code'].astype('category')
else:
    print("The 'Prog Code' column is not present in the DataFrame.")

# Group by 'Prog Code' and count the number of students in each program
prog_counts = df.groupby('Prog Code')['ID No'].count()

# Generate a color palette with enough colors for each group
num_groups = len(prog_counts)
colors = sns.color_palette('viridis', n_colors=num_groups)

# Create the bar graph
plt.figure(figsize=(10, 6))  # Adjust figure size as needed
bars = plt.bar(prog_counts.index, prog_counts.values, color=colors)
plt.xlabel('Program Code', fontsize=12)
plt.ylabel('Number of Students', fontsize=12)
plt.title('Number of Students by Program Code', fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=10)  # Rotate x-axis labels for readability
plt.yticks(fontsize=10)


# Adding labels to each bar
for bar in bars:
  yval = bar.get_height()
  plt.text(bar.get_x() + bar.get_width()/2, yval, int(yval), va='bottom', ha='center', fontsize = 8)

plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def plot_grouped_bars(df, column, group_by=None, bins=10):


    plt.figure(figsize=(10, 6))

    if group_by:
           grouped_counts = df.groupby(group_by)[column].count()
           num_groups = len(grouped_counts)
           colors = sns.color_palette('viridis', n_colors=num_groups)
           bars = plt.bar(grouped_counts.index, grouped_counts.values, color=colors)
           plt.xlabel(group_by, fontsize=12)
           plt.ylabel(f'Count of {column}', fontsize=12)
           plt.title(f'Distribution of {column} by {group_by}', fontsize=14)
           plt.xticks(rotation=45, ha='right', fontsize=10)

    else:
           if pd.api.types.is_numeric_dtype(df[column]):
               # Bin numerical data
               df['binned'] = pd.cut(df[column], bins=bins)
               binned_counts = df['binned'].value_counts().sort_index()
               colors = sns.color_palette('viridis', n_colors = len(binned_counts))
               bars = plt.bar(binned_counts.index.astype(str), binned_counts.values, color=colors)
               plt.xlabel(f'{column} Bins', fontsize=12)
               plt.ylabel('Frequency', fontsize=12)
               plt.title(f'Distribution of {column} (Binned)', fontsize=14)
               plt.xticks(rotation=45, ha='right', fontsize=10)
               del df['binned']
           else:
               # Count unique values
               counts = df[column].value_counts()
               num_groups = len(counts)
               colors = sns.color_palette('viridis', n_colors=num_groups)
               bars = plt.bar(counts.index, counts.values, color=colors)
               plt.xlabel(column, fontsize=12)
               plt.ylabel('Frequency', fontsize=12)
               plt.title(f'Distribution of {column}', fontsize=14)
               plt.xticks(rotation=45, ha='right', fontsize=10)

    plt.yticks(fontsize=10)
    for bar in bars:
          yval = bar.get_height()
          plt.text(bar.get_x() + bar.get_width()/2, yval, int(yval), va='bottom', ha='center', fontsize = 8)

    plt.tight_layout()
    plt.show()
```
